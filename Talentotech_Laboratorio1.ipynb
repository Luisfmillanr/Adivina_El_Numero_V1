```python
from google.colab import files
import numpy as np
import pandas as pd
import re

# Subir el archivo
uploaded = files.upload()

# Cargar el archivo
file_name = 'datalimpiarla.xlsx'  # Asegúrate de usar el nombre correcto
data = pd.read_excel(file_name)

# 1. Verificar y corregir duplicados en 'id'
if 'id' in data.columns:
    data['id'] = pd.to_numeric(data['id'], errors='coerce')  # Convertir a numérico, NaN si no es válido
    data = data.drop_duplicates(subset='id', keep='first')  # Eliminar filas con 'id' duplicados completamente

# 2. Corrección de 'cedula'
if 'cedula' in data.columns:
    data['cedula'] = data['cedula'].astype(str).str.replace(r'\D', '', regex=True)  # Remover caracteres no numéricos
    data['cedula'] = pd.to_numeric(data['cedula'], errors='coerce')  # Convertir a numérico, NaN si no es válido

# 3. Normalización de 'nombres y apellidos'
if 'nombres y apellidos' in data.columns:
    data['nombres y apellidos'] = data['nombres y apellidos'].str.strip().str.title().replace(r'\s+', ' ', regex=True)

# 4. Estandarización de 'direccion'
if 'direccion' in data.columns:
    data['direccion'] = data['direccion'].str.lower().str.strip().str.replace(r'cr\s*|cl\s*', lambda x: x.group().lower(), regex=True)

# 5. Normalización de 'correo electronico'
if 'correo electronico' in data.columns:
    data['correo electronico'] = data['correo electronico'].str.lower().str.strip()
    data['correo electronico'] = data['correo electronico'].replace(r'[^a-zA-Z0-9@._-]', '', regex=True)

# 6. Normalización y corrección de 'pais'
if 'pais' in data.columns:
    data['pais'] = data['pais'].str.lower().str.strip()
    country_corrections = {
        'colombiaaa': 'colombia',
        'chiles': 'chile',
        'japonn': 'japon',
        'rapon': 'japon',
        'pero': 'peru'
        # Agrega más correcciones según sea necesario
    }
    data['pais'] = data['pais'].replace(country_corrections)

# 7. Normalización de 'servicio' con eliminación de espacios adicionales
if 'servicio' in data.columns:
    data['servicio'] = data['servicio'].str.lower().str.strip().replace(r'\s+', ' ', regex=True)
    service_corrections = {
        'hogarr': 'hogar',
        'rogar': 'hogar',
        'comerr': 'comercio',
        'comercioo': 'comercio',
        'comerciio': 'comercio',
        'comerciooo': 'comercio',
        'loguearrr': 'comercio'
    }
    data['servicio'] = data['servicio'].replace(service_corrections)

# 8. Corrección de formato de 'fecha instalacion'
if 'fecha instalacion' in data.columns:
    data['fecha instalacion'] = pd.to_datetime(data['fecha instalacion'], errors='coerce')

# Verificación y revisión final de la limpieza
print("Primeras filas después de la limpieza:")
print(data.head())

# Exportar el DataFrame limpio si es necesario
data.to_excel('datalimpio_final.xlsx', index=False)
print("\nEl archivo limpio ha sido guardado como 'datalimpio_final.xlsx'.")

files.download('datalimpio_final.xlsx') #Por favor lean los comentarios explica los pasos si necesitan cambios con el chatgpt toma la seccion y explica el cambio que quiere hacer
```
